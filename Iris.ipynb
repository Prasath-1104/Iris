```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
# Load the Iris dataset
iris = load_iris()

# Convert to a pandas DataFrame for easier manipulation
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
```

### Iris Dataset Overview

```python
# Display the first 5 rows of the dataset
display(iris_df.head())
```

```python
# Display basic information about the dataset
display(iris_df.info())
```

```python
# Map the species column to actual names for better understanding
species_names = iris.target_names
iris_df['species_name'] = iris_df['species'].map(lambda x: species_names[x])

display(iris_df.head())
```

### Data Visualization: Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pair plot to visualize relatAionships between features
sns.pairplot(iris_df, hue='species_name', palette='viridis')
plt.suptitle('Pair Plot of Iris Dataset Features', y=1.02) # Adjust title position
plt.show()
```
